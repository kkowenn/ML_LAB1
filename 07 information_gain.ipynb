```python
from math import log2

# Function to calculate entropy
def entropy(p):
    if p == 0 or p == 1:
        return 0
    return -(p * log2(p) + (1 - p) * log2(1 - p))

# Given entropies for each attribute
entropy_shape_round = 0.72
entropy_shape_oval = 0.72
entropy_taste_sweet = 0.98
entropy_taste_sour = 0.92
entropy_color_orange = 0.81
entropy_color_green = 0.92

# Given data
total_samples = 10
samples_orange = 5
samples_mango = 5

# Root entropy
entropy_root = entropy(samples_orange / total_samples)

# Entropy after splitting by shape
# We need the number of round and oval samples
# From the given dataset image, we count the instances
round_samples = 4
oval_samples = 6

# Weighted entropy for shape
weighted_entropy_shape = (round_samples / total_samples) * entropy_shape_round + \
                         (oval_samples / total_samples) * entropy_shape_oval

# Entropy after splitting by taste
# Count the instances of sweet and sour taste from the given dataset
sweet_samples = 6
sour_samples = 4

# Weighted entropy for taste
weighted_entropy_taste = (sweet_samples / total_samples) * entropy_taste_sweet + \
                         (sour_samples / total_samples) * entropy_taste_sour

# Entropy after splitting by color
# Count the instances of orange and green color from the given dataset
orange_samples = 5
green_samples = 5

# Weighted entropy for color
weighted_entropy_color = (orange_samples / total_samples) * entropy_color_orange + \
                         (green_samples / total_samples) * entropy_color_green

# Calculate Information Gain for each attribute
information_gain_shape = entropy_root - weighted_entropy_shape
information_gain_taste = entropy_root - weighted_entropy_taste
information_gain_color = entropy_root - weighted_entropy_color

{
    "Entropy Root": entropy_root,
    "Weighted Entropy Shape": weighted_entropy_shape,
    "Weighted Entropy Taste": weighted_entropy_taste,
    "Weighted Entropy Color": weighted_entropy_color,
    "Information Gain Shape": information_gain_shape,
    "Information Gain Taste": information_gain_taste,
    "Information Gain Color": information_gain_color
}
```
