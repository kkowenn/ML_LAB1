```python
import pandas as pd
import random

def sigmoid(x):
    return 1 / (1 + (2.71828 ** -x))

def sigmoid_derivative(y):
    return y * (1 - y)

file_path = 'bmi.csv'
df = pd.read_csv(file_path)
num_rows = df.shape[0]
random_row_index = random.randint(0, num_rows - 1)

x1 = df.loc[random_row_index, 'Height']
x2 = df.loc[random_row_index, 'Weight']
Y1 = df.loc[random_row_index, 'Index']

# Initialize weights and biases for the hidden layer
w11, w12, w13 = 0.5, 0.9, 0.5
b1 = 0.1

w21, w22, w23 = 0.4, 1.0, 0.5
b2 = -0.2

w31, w32, w33 = -1.2, 1.1, 1.0
b3 = 0.5

alpha = 0.01

# Calculate the weighted sum for the hidden layer
h11 = x1 * w11 + x2 * w21 + b1
h12 = x1 * w12 + x2 * w22 + b2
h13 = x1 * w13 + x2 * w23 + b3

# Apply the sigmoid activation function to the hidden layer outputs
out_h11 = sigmoid(h11)
out_h12 = sigmoid(h12)
out_h13 = sigmoid(h13)

# Calculate the weighted sum for the output layer
net_o = out_h11 * w31 + out_h12 * w32 + out_h13 * w33 + b3
Yk_p = sigmoid(net_o)

# Error signal at the output layer
error = Y1 - Yk_p
F_prime_o = Yk_p * (1 - Yk_p)

# Update weights for the output layer
w31 = w31 + alpha * out_h11 * error
w32 = w32 + alpha * out_h12 * error
w33 = w33 + alpha * out_h13 * error

# Update biases for the output layer
b3 = b3 + alpha * F_prime_o * error

# Calculate the error gradient for each neuron in the hidden layer
delta_h11 = sigmoid_derivative(out_h11) * (w31 * F_prime_o * error)
delta_h12 = sigmoid_derivative(out_h12) * (w32 * F_prime_o * error)
delta_h13 = sigmoid_derivative(out_h13) * (w33 * F_prime_o * error)

# Update weights for the hidden layer
w11 = w11 + alpha * x1 * delta_h11
w12 = w12 + alpha * x1 * delta_h12
w13 = w13 + alpha * x1 * delta_h13

w21 = w21 + alpha * x2 * delta_h11
w22 = w22 + alpha * x2 * delta_h12
w23 = w23 + alpha * x2 * delta_h13

# Update biases for the hidden layer
b1 = b1 + alpha * delta_h11
b2 = b2 + alpha * delta_h12
b3 = b3 + alpha * delta_h13

# Print the calculated values and updated weights and biases
print(f'x1: {x1}, x2: {x2}, Y1: {Y1}')
print(f'Yk(p): {Yk_p}')
print(f'Updated weights for the output layer: w31: {w31}, w32: {w32}, w33: {w33}')
print(f'Updated biases for the output layer: b3: {b3}')
print(f'Error gradient at the hidden layer: delta_h11: {delta_h11}, delta_h12: {delta_h12}, delta_h13: {delta_h13}')
print(f'Updated weights for the hidden layer: w11: {w11}, w12: {w12}, w13: {w13}, w21: {w21}, w22: {w22}, w23: {w23}')
print(f'Updated biases for the hidden layer: b1: {b1}, b2: {b2}, b3: {b3}')
```
